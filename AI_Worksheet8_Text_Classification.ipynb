```python
def lower_case(text):
   return text.lower()

def remove_url(text):
  return re.sub(r'https?://\S+|www\.\S+','',text)

def remove_unwanted_char(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
        "]+",
        flags=re.UNICODE
    )

    text = emoji_pattern.sub('', text)

    # remove mentions & hashtags
    text = re.sub(r'[@#]\w+', '', text)

    # remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # FIX: keep spaces
    text = re.sub(r'\s+', ' ', text)

    return text.strip()
```

```python

def tokenize_stop_word(text):

    stop_words = set(stopwords.words('english'))
    stop_words.add('rt')

    tokenize_text = word_tokenize(text)

    tokenize_text = [token for token in tokenize_text if token.lower() not in stop_words]

    return tokenize_text

def lame_stem(tokens, lemm=True):

    if lemm:
        lemmatizer = WordNetLemmatizer()
        return [lemmatizer.lemmatize(token, pos='v') for token in tokens]

    else:
        stemmer = PorterStemmer()
        return [stemmer.stem(token) for token in tokens]
```

# Build a Text Cleaning Pipeline

```python
def text_cleaning_pipeline(text, rule="lemmatize"):
    """
    This function cleans text and applies NLP pipeline
    """

    # Convert to lowercase
    data = lower_case(text)

    # Remove URLs
    data = remove_url(data)

    # Remove emojis / unwanted chars
    data = remove_unwanted_char(data)

    # Create tokens
    tokens = data.split()

    # Remove stopwords
    tokens = tokenize_stop_word(data)

    # Lemmatize or Stem
    if rule == "lemmatize":
        tokens = lame_stem(tokens, lemm=True)

    elif rule == "stem":
        tokens = lame_stem(tokens, lemm=False)

    else:
        print("Pick between lemmatize or stem")

    return " ".join(tokens)
```

```python
data['clean_text']=data['text'].apply(text_cleaning_pipeline)
data
```

# Text Classification using Machine Learning Models


```python

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data['clean_text'],
    data['Sentiment'],
    test_size=0.2,
    random_state=42
)

print(f"X_train: {len(X_train)}, y_train: {len(y_train)}, X_test: {len(X_test)}, y_test: {len(y_test)}")
```

```python
# create vectorizer
vectorizer = TfidfVectorizer(max_features=500)

# fit on training data and transform
X_train_tfidf = vectorizer.fit_transform(X_train)

# transform test data
X_test_tfidf = vectorizer.transform(X_test)
```

```python
# Create model
model = LogisticRegression()

# Train model
model.fit(X_train_tfidf, y_train)

# Predict
y_pred = model.predict(X_test_tfidf)

# Evaluation
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
